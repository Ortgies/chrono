# Extracting text using image recognition
I use Tesseract to find text information in the product pictures of watch offerings. <br>
For every picture I extract a list of detected strings. These are later clustered to group similar words. <br>
<br>
To run my text recognition in multiple processes in Jupyter, I had to write my text recognition functions in a diffrent file called defs.py. <br>

```python
import os
import pandas as pd
import multiprocess as mp
import defs
```

```python
files = os.listdir('.\images')
len(files)
```

```python
if os.path.isfile('words.csv'):
    df = pd.read_csv('words.csv', low_memory=False)
    df.set_index('id', inplace=True)
else:
    df = pd.DataFrame()

print(len(df))
df.tail()
```

```python
while len(df) < len(files):
    start = len(df)
    end = start+120

    ids = [i.split('.')[0] for i in files[start:end]]

    if __name__ == '__main__':
        with mp.Pool(6) as p:
            x = p.map(defs.get_word_list, files[start:end])
            
    
    df_temp = pd.DataFrame(x)
    df_temp['id'] = ids
    df_temp.columns = df_temp.columns.map(str)
    df_temp.set_index('id', inplace=True)
    df = pd.concat([df, df_temp], axis=0)
    
    print(len(df))
```

```python
if os.path.isfile('words.csv'):
    
    if os.path.isfile('words_backup.csv'):
        os.remove('words_backup.csv')
    
    os.rename('words.csv', 'words_backup.csv')

df.to_csv('words.csv')
```
