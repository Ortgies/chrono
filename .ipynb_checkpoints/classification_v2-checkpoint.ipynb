```python
from fastai.vision.all import *
import pandas as pd
```

```python
fnames = os.listdir('/workspace/data/images/')
fnames = [re.split(r'.jpg$', x)[0] for x in fnames]
len(fnames)
```

```python
df = pd.read_csv('data_v2.csv', low_memory=False)
df = df[df['id'].isin(fnames)]
df['is_valid'] = df['is_valid'] == 1
df.reset_index(inplace=True, drop=True)
df.head()
```

```python
df['is_valid'].value_counts()
```

```python
def splitter(df):
    train = df.index[~df['is_valid']].tolist()
    valid = df.index[df['is_valid']].tolist()
    return train,valid
```

```python
def get_x(r): return '/workspace/data/images/{}.jpg'.format(r['id'])
def get_y(r): return r['Brand']
```

```python
get_x(df.loc[0]), get_y(df.loc[0])
```

```python
dblock = DataBlock(blocks=(ImageBlock, CategoryBlock),
                   splitter=splitter,
                   get_x=get_x, 
                   get_y=get_y,
                   item_tfms=Resize(460),
                   batch_tfms=Rotate(max_deg=20)
                  )
dls = dblock.dataloaders(df)
```

```python
dls.show_batch(nrows=3, ncols=5)
```

```python
torch.save(dls, 'images.pkl')
```
