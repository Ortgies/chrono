## Categorical features
Aside from the year of production and case diameter, all features are categorical. I set the data type to make sure everything works as intended later on.
For some categories I can set an internal order to make the decision tree more efficient. In the case of materials the order is a bit arbitrary, but precious metals should definitely be separated from steel and other affordable materials.

```python
def order_category(cat, order):
    df[cat] = df[cat].astype('category')
    df[cat] = df[cat].cat.set_categories(order, ordered=True)
```

```python
bracelet = 'Platinum','White gold','Rose gold','Yellow gold','Gold/Steel','Silver', 'Titanium', 'Ceramic', 'Steel', 'Aluminium', 'Leather', 'Textile', 'Rubber'
order_category('Bracelet material', bracelet)

case = 'Platinum', 'White gold', 'Rose gold', 'Yellow gold', 'Gold/Steel', 'Silver', 'Titanium', 'Ceramic', 'Steel', 'Bronze', 'Carbon','Plastic',  'Aluminum', 'Other'
order_category('Case material', case)

condition = 'New', 'Unworn', 'Very good', 'Good', 'Fair', 'Poor'
order_category('Condition', condition)

movement = 'Automatic', 'Manual winding', 'Quartz'
order_category('Movement', movement)

delivery = 'Original box, original papers', 'Original papers, no original box', 'Original box, no original papers', 'No original box, no original papers', 
order_category('Scope of delivery', delivery)
```

## Set up training and validation datasets

```python
def splitter(df):
    train = df.index[df['is_valid'] == 0].tolist()
    valid = df.index[df['is_valid'] == 1].tolist()
    return train,valid


to = TabularPandas(df.drop(columns=['is_valid']), [Categorify], cat, cont, y_names='price', splits=splitter(df))
len(to.train),len(to.valid)
```

## Feature importance
I try to slim down the model a bit by removing features that are redundant or have little to no information content.

```python
fi = pd.DataFrame({'cols':df.columns.drop(['price', 'is_valid']), 'imp':model.feature_importances_}).sort_values('imp', ascending=False)
fi.plot('cols', 'imp', 'barh', figsize=(12,7), legend=False)
```

```python
def cluster_columns(df, figsize=(10,6), font_size=12):
    corr = np.round(scipy.stats.spearmanr(df).correlation, 4)
    corr_condensed = hc.distance.squareform(1-corr)
    z = hc.linkage(corr_condensed, method='average')
    fig = plt.figure(figsize=figsize)
    hc.dendrogram(z, labels=df.columns, orientation='left', leaf_font_size=font_size)
    plt.show()
```

Not surprisingly, location and country of sale are redundant.

## Watefall plot
A waterfall plot is a nice way to visualize the regression logic on a case by case basis. <br>
Positive values indicate that a feature adds to the price, while negative values reduce price. <br>
The baseline is the bias, in this case the mean price of all watches. <br>
In the example below, the model would predict a log price of 7.72 (8.58 - 0.86), which is equivalent to $2245.

```python
waterfall(valid_xs_i.columns, contributions[1], threshold=0.05, 
          rotation_value=45,formatting='{:,.3f}')
```

```python
bias[1],bias[1]+np.sum(contributions[1]),np.exp(bias[1]+np.sum(contributions[1]))
```

The log price isnt very intuitive, so it might make sense to reverse the transformation and look at absolute prices. <br>
The price is again predicted by subtracting the sum of contributions (-$3062) from the mean price of 5320â‚¬.

```python
def absolute_contributions(bias, contributions):
    a = []
    b = bias  ## start with the bias as a baseline
    for n in contributions:
        b2 = b + n  ## add contribution to bias
        c = np.exp(b2) - np.exp(b) ## change in absolute terms
        a.append(c)
        b = b2 ## update baseline to new value
    return a
        

waterfall(valid_xs_i.columns, absolute_contributions(bias[1], contributions[1]), threshold=0.1, 
          rotation_value=45,formatting='{:,.0f}')
```

## Embeddings
The neural net generates embeddings from the categorical variables. <br>
It might be interesting to take a closer look how the features look like when converted to a continuous scale.

```python
from sklearn.decomposition import PCA
```

```python
#function to embed features ,obtained from fastai forums
def embed_features(learner, xs):
    xs = xs.copy()
    for i, feature in enumerate(learner.dls.cat_names):
        emb = learner.model.embeds[i]
        new_feat = pd.DataFrame(emb(tensor(xs[feature], dtype=torch.int64, device='cuda:0')), index=xs.index, columns=[f'{feature}_{j}' for j in range(emb.embedding_dim)])
        xs.drop(columns=feature, inplace=True)
        xs = xs.join(new_feat)
    return xs
```

```python
emb_xs = embed_features(learn, to_nn.train.xs)
```

To display the embeddings in a 2d plot, I reduce the dimensionality with principal component analysis.

```python
def get_feature_embs(feature, components, threshold):
    out = emb_xs.filter(regex=feature)
    values = df.loc[out.index.values, feature].values ## save the pre-encoded value names
    
    out = PCA(n_components=components).fit_transform(out)
    out = pd.DataFrame(out)
    out['value'] = values
    out = out.groupby('value').filter(lambda x : len(x)>(len(out)*threshold)) ## filter out values based on threshold
    out = out.groupby('value').first()
    out = out.dropna()
    
    return out
```

As a first example, lets look at the embeddings of case material. <br>
We can see two larger clusters, with precious metals on the right and more affordable materials on the left. The three types of gold are very close to each other. Platinum and Steel/Gold are two outliers and sit in opposite corners.

```python
emb_case = get_feature_embs('Case material', 2, 0.005)

plt.figure(figsize=(15,15))
plt.scatter(emb_case[0], emb_case[1])
for i, x, y in zip(emb_case.index, emb_case[0], emb_case[1]):
    plt.text(x+0.001,y,i, color=np.random.rand(3)*0.7, fontsize=10)
plt.show()
```

The most interesting embeddings are probably the Brands. <br>
The model does not know anything about the Brands itself but is still able to cluster very similar brands. Vacheron Constantin, Audemars Piguet and Patek Phillipe are known as the "holy trinity" of watchmakers and are located right next to each other in the 2d space. Jaeger LeCoultre and Breguet have similar prestige and are often considered as the fourth member of the "trinity", and they are right next to the others in the plot. <br>
In the top left corner are affordable fashion brands like Casio, Seiko or Tissot. <br>
Rolex is a very unusual brand and not really comparable to anything else, so it is no surprise that they sit very isolated in the embedding space.

```python
emb_brand = get_feature_embs('Brand', 2, 0.005)

plt.figure(figsize=(15,15))
plt.scatter(emb_brand[0], emb_brand[1])
for i, x, y in zip(emb_brand.index, emb_brand[0], emb_brand[1]):
    plt.text(x+0.002,y,i, color=np.random.rand(3)*0.7, fontsize=10)
plt.show()
```
